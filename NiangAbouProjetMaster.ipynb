```python
import nltk  # On a d'abord installé la bibliotheque NLTK qui est bibliotheque importante en traitement automatique de laangues
# puis on a proced a son importation
from collections import Counter  # importation de Counter pour compter les occurrences des mots
from nltk.tokenize import word_tokenize  # importation de la fonction de tokenisation des mots

# Télécharger les ressources nécessaires pour la tokenisation
nltk.download('punkt')  # Téléchargement de la ressource 'punkt' de NLTK, nécessaire pour le tokenization des mots
nltk.download('punkt_tab')  # Téléchargement de la ressource 'punkt_tab', si nécessaire
nltk.data.path.append(r"C:\Users\dell\AppData\Roaming\nltk_data")  # Ajout du chemin de la base de données NLTK local
```

```python
# Charger le fichier corpus
nom_fichier = "mon_corpus.txt"  # Spécifie le nom du fichier contenant le corpus à analyser
try:
    # Ouvrir le fichier en mode lecture avec encodage UTF-8
    with open(nom_fichier, "r", encoding="utf-8") as fichier:
        texte = fichier.read()  # Lire tout le contenu du fichier dans une variable 'texte'
except FileNotFoundError:
    # Si le fichier est introuvable, afficher un message d'erreur et quitter le programme
    print(f"Erreur : Le fichier '{nom_fichier}' est introuvable.")
    exit()

# Tokeniser le texte
try:
    # Utilisation de la fonction word_tokenize de NLTK pour découper le texte en mots (tokens)
    tokens = word_tokenize(texte)
except LookupError as e:
    # Si une erreur survient lors de la tokenisation (par exemple, les ressources nécessaires ne sont pas téléchargées)
    print("Erreur lors de la tokenisation. Assurez-vous que 'punkt' et 'punkt_tab' sont bien installés.")
    print(e)
    exit()

# Compter les fréquences
frequence = Counter(tokens)  # Utilisation de Counter pour compter les occurrences de chaque token dans le texte

# Chercher un terme spécifique
terme = "radiothérapie"  # Spécifie le terme à rechercher dans le corpus
nombre_occurences = frequence[terme]  # Compte le nombre d'occurrences du terme spécifié
print(f"Le terme '{terme}' apparaît {nombre_occurences} fois dans le corpus.")

# Compter les tokens uniques et totaux
print(f"Nombre de tokens uniques : {len(frequence)}")  # Affiche le nombre total de tokens uniques dans le corpus
print(f"Nombre total de tokens (incluant répétitions) : {sum(frequence.values())}")  # Affiche le nombre total de tokens incluant les répétitions
```

apres les résultat de cet algorithme, j'en deduit que mon corpus contient *78 664 tokens*.

on a utilisé une méthode statistique(n-grammes) pour repérer les candidats termes
ci-dessous l'algorithme

```python
#statistique
from nltk.tokenize import word_tokenize  # Importation de la fonction pour tokenizer du texte
from nltk.corpus import stopwords  # Importation de la liste des mots vides en français
from nltk.util import ngrams  # Importation la fonction pour créer des n-grammes
from collections import Counter  # Importation de Counter pour compter les occurrences


# Téléchargement des ressources nécessaires
nltk.download('punkt')  # Télécharge les ressources pour la tokenisation des mots
nltk.download('stopwords')  # Télécharge les mots vides en français nécessaires pour le filtrage

# Charger le fichier corpus
nom_fichier = "mon_corpus.txt"  # Déclare le nom du fichier contenant le texte à analyser
with open(nom_fichier, "r", encoding="utf-8") as fichier:
    texte = fichier.read()  # Lit le contenu du fichier dans la variable texte

# Tokeniser le texte
tokens = word_tokenize(texte)  # Tokenise le texte en mots individuels

# Charger la liste des mots vides en français
mots_vides = set(stopwords.words('french'))  # Charge les mots vides en français et les stocke dans un ensemble

# Filtrer les mots vides et ceux de petite longueur
# Filtre les mots pour n'inclure que ceux qui sont plus longs que 3 caractères, sans mots vides et sans caractères spéciaux
tokens_filtres = [mot.lower() for mot in tokens if mot.lower() not in mots_vides and len(mot) > 3 and mot.isalpha()]

# Créer des bigrammes à partir des tokens filtrés
# Utilise la fonction ngrams de NLTK pour créer des bigrammes (groupes de deux mots consécutifs) à partir des tokens filtrés
bigrammes = list(ngrams(tokens_filtres, 2))

# Compter les bigrammes fréquents
# Utilise Counter pour compter le nombre d'occurrences de chaque bigramme
frequence_bigrammes = Counter(bigrammes)

# Affiche les bigrammes les plus fréquents
# Affiche les 20 bigrammes les plus fréquents après filtrage des mots vides
print("Bigrammes candidats (sans mots vides) :")
for bigramme, count in frequence_bigrammes.most_common(20):
    print(f"{' '.join(bigramme)}: {count}")
```

```python
import networkx as nx  # Bibliothèque pour manipuler les graphes
import matplotlib.pyplot as plt  # Bibliothèque pour visualiser les graphes

# Étape 1 : Initialisation du graphe orienté
# On crée un graphe orienté pour représenter les relations directionnelles entre les concepts.
G = nx.DiGraph() # Itiatisation de notre graph

# Étape 2 : Ajout des nœuds (concepts principaux liés à la radiothérapie)
# Les nœuds représentent des concepts validés, chacun ayant une signification spécifique.
terms = [ #ici, on a utiliseé une liste pour nous faciliter la manipulation
    "Radiothérapie", "Techniques", "Irradiation", "Cancers",
    "Centres", "Recherche clinique", "Soins",
    "Modulation d’intensité", "Radiothérapie stéréotaxique"
]
G.add_nodes_from(terms)  # Ajout des termes comme nœuds dans le graphe

# Étape 3 : Ajout les relations (arêtes) entre les noeuds
# Chaque relation a une étiquette pour expliquer la nature du lien entre deux concepts.
relations = [
    ("Radiothérapie", "Techniques", "utilise des"),  # La radiothérapie utilise des techniques
    ("Radiothérapie", "Irradiation", "implique"),  # La radiothérapie implique une irradiation
    ("Radiothérapie", "Cancers", "cible les"),  # La radiothérapie cible les cancers
    ("Techniques", "Modulation d’intensité", "comprend"),  # Les techniques incluent la modulation d’intensité
    ("Techniques", "Radiothérapie stéréotaxique", "inclut"),  # Les techniques incluent aussi la radiothérapie stéréotaxique
    ("Cancers", "Centres", "traités dans des"),  # Les cancers sont traités dans des centres spécialisés
    ("Cancers", "Soins", "nécessitent des"),  # Les cancers nécessitent des soins adaptés
    ("Centres", "Contrôle qualité", "assurent le"),  # Les centres assurent le contrôle qualité
    ("Centres", "Recherche clinique", "sont impliqués dans la"),  # Les centres participent à la recherche clinique
    ("Recherche clinique", "Techniques", "évalue l’efficacité des"),  # La recherche évalue les techniques
    ("Recherche clinique", "Soins", "vise à améliorer les"),  # La recherche vise à améliorer les soins
    ("Modulation d’intensité", "Radiothérapie", "utilisées dans la"),  # La modulation est utilisée en radiothérapie
    ("Radiothérapie stéréotaxique", "Radiothérapie", "utilisées dans la"),  # Idem pour cette technique
    ("Modulation d’intensité", "Irradiation", "permettent de minimiser"),  # Réduction des dommages aux tissus sains
    ("Radiothérapie stéréotaxique", "Irradiation", "permettent de minimiser")  # Idem pour cette technique
]

# Ajout des relations au graphe avec leurs labels
for source, target, label in relations:
    G.add_edge(source, target, label=label)

# Étape 4 : Configuration de la mise en page du graphe
# Le layout "spring_layout" place les noeuds de manière naturelle, et `k=0.8` augmente l'espacement.
pos = nx.spring_layout(G, seed=42, k=0.8)

# Étape 5 : Dessin du graphe avec des paramètres ajustés pour une meilleure lisibilité
plt.figure(figsize=(20, 15))  # Taille de la figure étendue pour une vue claire
nx.draw(
    G, pos, with_labels=True, node_color="lightblue",  # Couleur des nœuds
    node_size=4500, font_size=14, font_weight="bold",  # Taille et poids des labels des nœuds
    arrowsize=25, arrowstyle="->", edge_color="black"  # Taille et style des flèches
)

# Étape 6 : Ajouter les labels des relations
# Les étiquettes des relations (arêtes) sont ajoutées dans une couleur distincte et avec une taille lisible.
edge_labels = nx.get_edge_attributes(G, 'label')  # Récupérer les étiquettes des relations
nx.draw_networkx_edge_labels(
    G, pos, edge_labels=edge_labels, font_size=12, font_color="darkgreen"
)

# Étape 7 : Ajouter un titre au graphe
# Le titre donne un contexte au graphe pour le lecteur.
plt.title("Graphe de connaissances : Relations en radiothérapie", fontsize=18, fontweight="bold")

# Étape 8 : Afficher le graphe
plt.show()
```

```python
for source, target, label in relations:
    G.add_edge(source, target, label=label)
```

**Étape 5** : Configuration de la mise en page du graphe        
Pour rendre le graphe lisible et esthétique, j'utilise le layout "spring_layout" de NetworkX.       
Ce layout positionne les nœuds de manière naturelle en simulant une force de ressort entre eux.         
Le paramètre k contrôle l'espacement entre les nœuds (ici k=1.5 pour plus d'espacement) et seed=42 garantit une reproductibilité des positions.     

```python
pos = nx.spring_layout(G, seed=42, k=1.5)
```

**Étape 6** : Dessin du graphe avec des paramètres ajustés      
J'utilise Matplotlib pour dessiner le graphe en configurant des paramètres visuels pour améliorer la lisibilité :       

**figsize** : définit une taille étendue de la figure (20x15) pour éviter les chevauchements.       
**node_color** : couleur des nœuds pour les distinguer.     
**node_size et font_size** : ajustés pour rendre les nœuds et leurs labels visibles.        
**arrowsize et arrowstyle** : personnalisent les flèches représentant les relations.        

**Étape 7** : Ajout des labels des relations        
Les étiquettes des relations (ou labels des arêtes) sont récupérées avec get_edge_attributes et ajoutées au graphe.     
 Pour les différencier visuellement, j'utilise une couleur distincte (darkgreen) et une taille de police lisible.

**Étape 8** : Ajout d’un titre au graphe    
J’ajoute un titre pour donner un contexte clair au graphe et indiquer aux lecteurs son objectif.        
 Le titre est mis en forme avec une taille de police augmentée et un poids en gras.     

**Étape 9** : Affichage du graphe       
Enfin, j'utilise plt.show() pour afficher le graphe final       

```python
#Étape 6
plt.figure(figsize=(20, 15))  # Taille de la figure
nx.draw(
    G, pos, with_labels=True, node_color="lightblue",
    node_size=4500, font_size=14, font_weight="bold",
    arrowsize=25, arrowstyle="->", edge_color="black"
)

#Étape 7
edge_labels = nx.get_edge_attributes(G, 'label')  # Récupérer les labels des arêtes
nx.draw_networkx_edge_labels(
    G, pos, edge_labels=edge_labels, font_size=14, font_color="darkgreen"
)

#  Étape 8 : Ajout d’un titre au graphe 
plt.title("Graphe de connaissances : Relations en radiothérapie", fontsize=18, fontweight="bold")

plt.show()
```
